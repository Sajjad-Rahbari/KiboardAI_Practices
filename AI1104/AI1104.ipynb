```python
from cProfile import label

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ds_salaries.csv")
```

```python
df
```

## Question 1(only Answer):

```python
# Question 1
jop = (df[["job_title"]].groupby('job_title').size()).sort_values(ascending=False)
print(jop)
jobbest = (df[["job_title"]].groupby('job_title').size().sort_values(ascending=False))[:10]
plt.bar(jobbest.index,jobbest)
fig = plt.gcf()
fig.set_size_inches(32,8)
plt.xlabel("Job Title")
plt.ylabel("Salary in use")
plt.show()
```

## Question 2(only Answer):

```python
# Q2

plt.boxplot(list(df[df["work_year"]==2023]["salary_in_usd"]))
plt.xlabel("year 2023")
plt.ylabel("Salary in use")
plt.gcf().set_size_inches(10,10)
plt.show()
```

## Question 3(only Answer):

```python
# Q3

plt.hist(list(df[df["work_year"]==2023]["salary_in_usd"]))
plt.xlabel("Salary in use")
plt.gcf().set_size_inches(10,10)
plt.show()
```

## Question 4(only Answer):

```python
# Q4
listCountry = ("2023", "2022","2021", "2020") 
listSalary =(df[df["work_year"]==2023]["salary_in_usd"].mean(),df[df["work_year"]==2022]["salary_in_usd"].mean(),df[df["work_year"]==2021]["salary_in_usd"].mean(),df[df["work_year"]==2020]["salary_in_usd"].mean())
plt.bar(listCountry,listSalary)
plt.xlabel("YEAR")
plt.ylabel("Salary in use")
plt.show()
 
```

## Question 5(only Answer):

```python
# Q5

listCountry = ("EN", "MI", "SE", "EX")
listSalary = (df[df["experience_level"] == "EN"]["salary_in_usd"].mean(),
              df[df["experience_level"] == "MI"]["salary_in_usd"].mean(),
              df[df["experience_level"] == "SE"]["salary_in_usd"].mean(),
              df[df["experience_level"] == "EX"]["salary_in_usd"].mean())
plt.bar(listCountry, listSalary)
plt.xlabel("experience_level")
plt.ylabel("Salary in use")
plt.show()
 
```
