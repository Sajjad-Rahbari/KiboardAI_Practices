```python
# Q3

region = list(df6.groupby("Region").sum().index)
happyScoreOfRegion = list()
for i in region:
    happyScoreOfRegion.append(df6[df6["Region"]==i]["Happiness Score"])
plt.boxplot(happyScoreOfRegion,tick_labels=region)
plt.xlabel("Region")
plt.ylabel("Happiness Score ")
plt.gcf().set_size_inches(36,10)
plt.show()
```

## Question 4(only Answer):

```python
# Q3
df6.dropna(inplace=True)
iranH = df6[df6["Country"]=="Iran"]["Happiness Score"]
turkeyH = df6[df6["Country"]=="Turkey"]["Happiness Score"]
iraqH = df6[df6["Country"]=="Iraq"]["Happiness Score"]
germanyH = df6[df6["Country"]=="Germany"]["Happiness Score"]
kenyaH = df6[df6["Country"]=="Kenya"]["Happiness Score"]
worldH = (df6["Happiness Score"].sum() - iranH) / (len(df6["Happiness Score"]) - 1)
regionIranH = (df6[df6["Region"]=="Middle East and Northern Africa"]["Happiness Score"].sum() - iranH)/ (len(df6[df6["Region"]=="Middle East and Northern Africa"]) - 1)
hap = [float(iranH),float(turkeyH),float(iraqH),float(germanyH),float(kenyaH),float(regionIranH),float(worldH)]
names = ["iran","turkey","iraq","germany","kenya","Middle East\n and \nNorthern Africa","world"]
plt.bar(names,hap)
plt.xlabel("country and region and world")
plt.ylabel("happiness score")
plt.title("happiness score of Iran")
plt.gcf().set_size_inches(10,10)
plt.show()
```
