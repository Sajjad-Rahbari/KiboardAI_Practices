```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io

image = io.imread("Picture1.jpg", as_gray=True)

rows, cols = image.shape
arrayx = np.zeros((rows, cols*2))
```

```python
image
```

```python

plt.imshow(image)
```

## change image

```python
for i in range(cols):
    arrayx[:,i*2] = image[:,i]
```

```python
plt.imshow(arrayx)
```

## cut image to first (327*184)

```python
result = arrayx[:,:cols]
```
